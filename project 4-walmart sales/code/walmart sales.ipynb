## using statistical methods

```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

outlier_counts = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]
    
    print(f"Outliers for {col}:\n{outliers}\n")

print(f"Total number of outliers per column:\n{outlier_counts}")
```

```python
#there are 36 outliers outof 1000 rows in the data
```

### handling the outliers

```python
print(df.columns)
```

```python
# Assuming df is your dataframe
columns_to_fix = ['Tax 5%', 'Total', 'cogs', 'gross income']

outlier_counts_before = {}
outlier_counts_after = {}

for col in columns_to_fix:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count outliers before replacement
    outlier_counts_before[col] = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    
    # Calculate median
    median = df[col].median()
    
    # Replace outliers with median
    df[col] = df[col].apply(lambda x: median if (x < lower_bound or x > upper_bound) else x)
    
    # Count outliers after replacement
    outlier_counts_after[col] = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()

# Display counts
print("Outliers count before replacement:")
for col, count in outlier_counts_before.items():
    print(f"{col}: {count}")

print("\nOutliers count after replacement:")
for col, count in outlier_counts_after.items():
    print(f"{col}: {count}")
```

```python
import mysql.connector


# Database connection details
db_config = {
    'host': 'localhost',
    'user': 'root',        # Replace with your MySQL username
    'password': 'root',    # Replace with your MySQL password
    'database': 'projects' # Replace with your database name
}

# Connect to the database
connection = mysql.connector.connect(**db_config)
cursor = connection.cursor()

# Create the table if it doesn't exist
table_name = 'Walmart_sales_data'
create_table_query = f"""
CREATE TABLE IF NOT EXISTS {table_name} (
    Invoice_ID VARCHAR(50),
    Branch VARCHAR(10),
    City VARCHAR(100),
    Customer_Type VARCHAR(50),
    Gender VARCHAR(10),
    Product_Line VARCHAR(100),
    Unit_Price FLOAT,
    Quantity INT,
    Tax_five_per FLOAT,
    Total FLOAT,
    Date DATE,
    Payment VARCHAR(50),
    cogs FLOAT,
    Gross_Margin_Percentage FLOAT,
    Gross_Income FLOAT,
    Rating FLOAT
);
"""
cursor.execute(create_table_query)

# Filter required columns from DataFrame (assuming df is already defined)
df_to_insert = df[['Invoice_ID', 'Branch', 'City', 'Customer_type', 'Gender',
                   'Product_line', 'Unit_price', 'Quantity', 'Tax_five_per', 'Total',
                   'Date', 'Payment', 'cogs', 'gross_margin_percentage', 'gross_income', 'Rating']]

# Convert filtered DataFrame to list of tuples
data_to_insert = df_to_insert.values.tolist()


# Insert multiple rows at once
insert_query = f"""
INSERT INTO {table_name} (Invoice_ID, Branch, City, Customer_Type, Gender, Product_Line,
                          Unit_Price, Quantity,Tax_five_per, Total, Date, Payment,
                          cogs, Gross_Margin_Percentage, Gross_Income, Rating)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Insert data into the table
cursor.executemany(insert_query, data_to_insert)

# Commit the transaction and close the connection
connection.commit()
print(f"Data inserted into table '{table_name}' successfully.")

# Close the cursor and connection
cursor.close()
connection.close()
```
